```python
#one hot encoding for columns InternetService,Contract and PaymentMethod
telecom1 = pd.get_dummies(data = df, columns = ['InternetService','Contract','PaymentMethod'])
```

```python
#creating predictor and target variables
X = telecom1.drop('Churn',axis = 'columns')
Y = telecom1['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
#using pearson correlation
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sns.heatmap(X_train.corr(), cmap='YlGnBu',annot=True)
plt.show()
```

```python
#using pearson correlation feature selection funtion
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i,j] > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train,0.8)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train = X_train.drop(corr_features,axis=1)
X_test = X_test.drop(corr_features,axis=1)
```

```python
X_train.shape,X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
model = Sequential()
model.add(Dense(units=20, kernel_initializer='he_uniform',
                bias_initializer='ones', activation='tanh', input_dim=25))
model.add(Dense(units=10, kernel_initializer='he_uniform',
                bias_initializer='ones', activation='tanh'))
model.add(Dense(units=5, kernel_initializer='he_uniform',
                bias_initializer='ones', activation='tanh'))
model.add(Dense(units=3, kernel_initializer='he_uniform',
                bias_initializer='ones', activation='tanh'))
model.add(Dense(units=2, kernel_initializer='he_uniform',
                bias_initializer='ones', activation='tanh'))
model.add(Dense(units=1,  kernel_initializer='he_uniform',
                bias_initializer='ones', activation='sigmoid'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.001, amsgrad=False),loss='binary_crossentropy',metrics=['accuracy','mae'])
```

```python
#fitting the model
from keras.models import load_model
from keras.callbacks import EarlyStopping, ModelCheckpoint
telecom_classify = model
#create callbacks for optimized training
# callbacks = [EarlyStopping(monitor='val_loss', patience=2),
#              ModelCheckpoint(filepath='telecom_model.h5', monitor='val_loss', save_best_only=True)]
telecomm_built_model=telecom_classify.fit(X_train,Y_train, batch_size=32 ,epochs=100,validation_data=(X_test,Y_test),shuffle=False)
```

```python
model.save('./SiameseLSTM_new25.h5')
```

```python
print(str(telecomm_built_model.history['accuracy'][-1])[:6] +
      "(max of accuracy out of all the epochs trained: " + str(max(telecomm_built_model.history['accuracy']))[:6] + ")")

print(str(telecomm_built_model.history['val_accuracy'][-1])[:6] +
      "(max of val_accuracy of all the epochs trained: " + str(max(telecomm_built_model.history['val_accuracy']))[:6] + ")")
print("Done.")
```

```python
#plotting
print(telecomm_built_model.history.keys())
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(telecomm_built_model.history['accuracy'])
plt.plot(telecomm_built_model.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'Validation'], loc='upper left')
plt.show()
plt.savefig('./Accuracy.png')
```

```python
# summarize history for loss
plt.plot(telecomm_built_model.history['loss'])
plt.plot(telecomm_built_model.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'Validation'], loc='upper left')
plt.show()
plt.savefig('./Loss.png')
```

```python
predict_churn = model.predict(X_test)
y_pred = []
for i in predict_churn:
    if i > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import mean_absolute_error
print(classification_report(Y_test,y_pred))
```

```python
import seaborn as sns
confusion_matrix = tf.math.confusion_matrix(labels=Y_test,predictions=y_pred)
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix, fmt='d',annot=True,cmap='GnBu')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
accuracy = (confusion_matrix[0][0]+confusion_matrix[1][1])/(confusion_matrix[0][0]+confusion_matrix[0][1]+confusion_matrix[1][0]+confusion_matrix[1][1])
print("Confusion Matrix Accuracy: "+ str(accuracy)+"%")
```
